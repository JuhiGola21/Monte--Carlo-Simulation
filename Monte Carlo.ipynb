```python
# Step 1: Install required libraries (run this once)
!pip install numpy matplotlib

# Step 2: Import packages
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Set up parameters
initial_price = 100  # Starting stock price
days = 252           # 1 year of trading days
mu = 0.08            # Expected annual return (8%)
sigma = 0.20         # Annual volatility (20%)
num_paths = 20       # Number of random paths to generate

# Step 4: Generate random daily returns
daily_returns = np.random.normal(
    loc=mu/252,       # Daily mean return
    scale=sigma/np.sqrt(252),  # Daily volatility
    size=(days, num_paths)
)

# Step 5: Calculate price paths
price_paths = initial_price * (1 + daily_returns).cumprod(axis=0)

# Step 6: Plot results
plt.figure(figsize=(10, 6))
plt.plot(price_paths, linewidth=1)
plt.title(f"{num_paths} Random Stock Price Paths\n(μ={mu*100:.0f}%, σ={sigma*100:.0f}%)")
plt.xlabel("Trading Days")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()

# Step 7: Show final price statistics
final_prices = price_paths[-1, :]
print(f"\nAfter {days} days:")
print(f"Average price: ${np.mean(final_prices):.2f}")
print(f"Highest price: ${np.max(final_prices):.2f}")
print(f"Lowest price:  ${np.min(final_prices):.2f}")
```
